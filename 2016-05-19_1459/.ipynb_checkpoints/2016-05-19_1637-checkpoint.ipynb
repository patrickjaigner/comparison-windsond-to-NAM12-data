```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

```python
#windsond csv
folder = r'/Users/Pat/Documents/Harvard/Compare/2016-05-19_1637/'
filename = '2016-05-19_1637_windsond_data.txt'
ground_height = 213

df_ws = pd.read_csv(folder+filename, header = [0,1], delim_whitespace = 1)
df_ws
```

```python
#nam12 profiles 18pm
df_p0 = pd.read_table(folder+ 'Profiles/18/profile__0m.txt',sep=r'\s*', 
                      header = [11,12])


df_p1 = pd.read_table(folder + 'Profiles/18/profile__1800m.txt',sep=r'\s*', 
                      header = [11,12])




#build a gps true nam12 profile df_p
df_p0_mod = df_p0.loc[lambda x: df_p0.HGTS.m<1700,:]
df_p1_mod = df_p1.loc[lambda x: (df_p0.HGTS.m >= 1700) & (df_p0.HGTS.m <= 3900),:]
df_p2_mod = df_p0.loc[lambda x: df_p0.HGTS.m>3900,:]
df_p = df_p0_mod.append(df_p1_mod).append(df_p2_mod)

#calculate wind dir
df_p['wind-dir'] = pd.Series(270 - (np.arctan2(df_p['VWND']['S->N'],df_p['UWND']['W->E']) * (180 / np.pi)), index = df_p.index)
df_p['wind-dir'] = df_p['wind-dir'].map(lambda x: x-360 if x>360 else x)

#filter to max 5km
df_p = df_p.loc[lambda x: df_p.HGTS.m<5001,:]
df_p



```

```python
#nam12 profiles 15pm
df_p0_2 = pd.read_table(folder+ 'Profiles/15/profile__0m.txt',sep=r'\s*', 
                      header = [11,12])


df_p1_2 = pd.read_table(folder + 'Profiles/15/profile__1800m.txt',sep=r'\s*', 
                      header = [11,12])




#build a gps true nam12 profile df_p
df_p0_2_mod = df_p0_2.loc[lambda x: df_p0_2.HGTS.m<1700,:]
df_p1_2_mod = df_p1_2.loc[lambda x: (df_p0_2.HGTS.m >= 1700) & (df_p0_2.HGTS.m <= 3900),:]
df_p2_2_mod = df_p0_2.loc[lambda x: df_p0_2.HGTS.m>3900,:]
df_p_2 = df_p0_2_mod.append(df_p1_2_mod).append(df_p2_2_mod)

#calculate wind dir
df_p_2['wind-dir'] = pd.Series(270 - (np.arctan2(df_p_2['VWND']['S->N'],df_p_2['UWND']['W->E']) * (180 / np.pi)), index = df_p_2.index)
df_p_2['wind-dir'] = df_p_2['wind-dir'].map(lambda x: x-360 if x>360 else x)

#filter to max 5km
df_p_2 = df_p_2.loc[lambda x: df_p_2.HGTS.m<5001,:]
df_p_2
```

```python
from matplotlib.backends.backend_pdf import PdfPages

#do change
pp = PdfPages(folder + '2016-05-19_1637_nam12+18h_windsond.pdf')
nam12_h_offset = '+18h'
nam12_h_offset_2 = '+15h'
plot_title = 'fallCreek_19052016_1637'


#x = temp, y = height
plt.clf()
plt.figure(1)
plt.plot(df_p['TEMP']['oC'],df_p['HGTS'],'b',label = 'nam12'+nam12_h_offset)
plt.plot(df_p_2['TEMP']['oC'],df_p_2['HGTS'],'g',label = 'nam12'+nam12_h_offset_2)
plt.plot(df_ws['temp'],df_ws['alt']+ground_height,'r', label = 'windsond')
plt.ylabel('Height [m]')
plt.xlabel('Temp [C]')
plt.title('TEMP: '+plot_title)
plt.grid(True)
plt.legend(loc='best')
plt.savefig(pp, format='pdf')
#plt.show()

#x = temp, y = windspeed
plt.figure(2)
#plt.plot(df_p['UWND']['m/s'],df_p['HGTS'],'b',label = 'nam12 UWND')
windspd = np.sqrt((df_p['VWND']['m/s']*df_p['VWND']['m/s']) + (df_p['UWND']['m/s']*df_p['UWND']['m/s']))
plt.plot(windspd,df_p['HGTS'],'b',label = 'nam12'+nam12_h_offset)
windspd_2 = np.sqrt((df_p_2['VWND']['m/s']*df_p_2['VWND']['m/s']) + (df_p_2['UWND']['m/s']*df_p_2['UWND']['m/s']))
plt.plot(windspd_2,df_p_2['HGTS'],'g',label = 'nam12'+nam12_h_offset_2)
plt.plot(df_ws['spd'],df_ws['alt']+ground_height,'r', label = 'windsond')
plt.ylabel('Height [m]')
plt.xlabel('Windspeed [m/s]')
plt.title('WINDSPEED: '+plot_title)
plt.grid(True)
plt.legend(loc='best')
pp.savefig()
#plt.show()

#x = hum, y = height
plt.figure(3)
plt.plot(df_p['RELH'],df_p['HGTS'],'b',label = 'nam12'+nam12_h_offset)
plt.plot(df_p_2['RELH'],df_p_2['HGTS'],'g',label = 'nam12'+nam12_h_offset_2)
plt.plot(df_ws['hum'],df_ws['alt']+ground_height,'r', label = 'windsond')
plt.ylabel('Height [m]')
plt.xlabel('Humidity [%]')
plt.title('HUM: '+plot_title)
plt.grid(True)
plt.legend(loc='best')
pp.savefig()
#plt.show()


#x = wind-dir, y = height

plt.figure(4)
plt.plot(df_p['wind-dir'],df_p['HGTS'],'b',label = 'nam12'+nam12_h_offset)
plt.plot(df_p_2['wind-dir'],df_p_2['HGTS'],'g',label = 'nam12'+nam12_h_offset_2)
plt.plot(df_ws['wind-dir'],df_ws['alt']+ground_height,'r', label = 'windsond')
plt.ylabel('Height [m]')
plt.xlabel('Wind direction [o]')
plt.title('WIND-DIR: '+plot_title)
plt.grid(True)
plt.legend(loc='best')
pp.savefig()

pp.close()            

plt.show()
```
